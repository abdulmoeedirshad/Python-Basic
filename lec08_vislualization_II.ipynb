### Step 1: Place a kernel at each point

We'll start with the Gaussian kernel.

```python
def gaussian(x, z, a):
    
    return (1/np.sqrt(2*np.pi*a**2)) * np.exp((-(x - z)**2) / (2 * a**2))

points = [2.2, 2.8, 3.7, 5.3, 5.7]
a = 0.8

x_values = np.linspace(min(points)-5, max(points)+5, 100)

kernel_values = [gaussian(x_values, pt, a) for pt in points]

for i, pt in enumerate(points):
    plt.plot(x_values, kernel_values[i], label=f'Point {i+1} at x={pt}')

plt.xlabel('X_Values')
plt.ylabel('Y_Values')
plt.title('Gaussian Kernel for Given Points')
plt.legend()
plt.show()
```

```python
points = [2.2, 2.8, 3.7, 5.3, 5.7]
a = 0.8
n=len(points)

x_values = np.linspace(min(points)-5, max(points)+5, 100)

kernel_values = [gaussian(x_values, pt, a) for pt in points]

for i, pt in enumerate(points):
    plt.plot(x_values, kernel_values[i]/n, label=f'Point {i+1} at x={pt}')
plt.xlim(-3,10)
plt.ylim(0,0.5)
plt.xlabel('X_Values')
plt.ylabel('Y_Values')
plt.title('Gaussian Kernel for Given Points')
plt.legend()
plt.show()
```

### Step 3: Sum all kernels together

```python
# code here
```

```python
points = [2.2, 2.8, 3.7, 5.3, 5.7]
a = 0.8


x_values = np.linspace(min(points) - 5, max(points) + 5, 100)



summed_kernel = np.zeros_like(x_values)


for pt in points:
    kernel_values = np.array([gaussian(xi, pt, a) for xi in x_values])
    kernel_values=kernel_values/5
    summed_kernel += kernel_values

plt.plot(x_values, summed_kernel, label='Summed Kernel')

plt.xlim(-3,10)
plt.ylim(0,0.5)
plt.xlabel('X_Values')
plt.ylabel('Summed Y_Values')
plt.title('Sum of Gaussian Kernels for Given Points')
plt.legend()
plt.show()
```

			\frac{1}{\alpha}, \: \: \: |x - x_i| \leq \frac{\alpha}{2}\\
